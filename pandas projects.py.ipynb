```python
data={'names':['ali','omar','sara','lina'],
     'age':[25,22,28,30]}
labels=range(1,5)
df=pd.DataFrame(data,index=labels)
df
```

```python
average_age=df['age'].mean()
average_age
```

```python
data={'students':['ali','sara','omar','lina'],
     'scores':[85,90,85,95]}
df=pd.DataFrame(data)
df
unique_score=df['scores'].nunique()
unique_score
```

```python
data={'product':['phone','labtop','playstation'],
     'sales':[1500,800,500]}
df=pd.DataFrame(data,index=range(1,4))
df.rename(columns={'product':'item'},inplace=True)
df
```

```python
data={'products':['labtop','phone','playstation'],
     'sales':[1500, None,800]}
df=pd.DataFrame(data ,index=range(1,4))
df['sales'].fillna(df['sales'].mean())
df
```

```python
data={'students':['lina','aseel','sama','yousef'],
     'scores':[75,95,84,98]}
df=pd.DataFrame(data ,index=range(1,5))

max_score=df['scores'].max()
max_score
```

```python
data={'Employees':['sama','lina','jana','youssef','lamees'],
     'Age':[23,28,26,32,29],
     'Department':['HR','finance','HR','finance','finance']}
df=pd.DataFrame(data,index=range(1,6))
average_by_age=df.groupby('Department')['Age'].mean()
print(df)
average_by_age
```

```python
data={'Days':['monday','tuesday','wednsday','thurusday'],
     'Sales':[200,500,357,480]}
df=pd.DataFrame(data,index=range(1,5))
best_sales_day=df.loc[df['Sales'].idxmax(),'Days']
print(df)
best_sales_day
```

```python
data={'employees':['omar','lina','ahmed'],
     'start_date':['1-1-2023','12-10-2023','6-24-2023']}
df=pd.DataFrame(data,index=range(1,4))
print(df)
pd.to_datetime(df['start_date'])
```

```python
data={'students':['ali','sara','lina','ali'],
     'scores':[85,90,87,85]}
df=pd.DataFrame(data,index=range(1,5))
print(df)
remove_duplicate=df.drop_duplicates(df)
print(remove_duplicate)
```

```python
data={'names':['ali','lina','omar'],
     'age':[25,60,15]}
df=pd.DataFrame(data,index=range(1,4))

df['age_groups']=pd.cut(df['age'],bins=[0,18,40,100],labels=['young','adult','geriateric'])
print(df)
```
